```python
import googlemaps
import pandas as pd

# API

gmaps = googlemaps.Client(key='')
```

```python
def get_place_info(place):
    try:
        geocode_result = gmaps.geocode(place)
    except Exception as e:
        print(f"Error al obtener la información para el sitio '{place}': {e}")
        return None

    if geocode_result:
        # Extrae la información relevante del primer resultado (asumiendo que es el más relevante)
        location = geocode_result[0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']
        address = geocode_result[0]['formatted_address']

        return latitude, longitude, address
    else:
        print(f"No se encontró información para el sitio '{place}'.")
        return None, None, None

ing"
data = pd.read_excel("data_google.xlsx")


data['Latitud'] = None
data['Longitud'] = None
data['Dirección'] = None


for index, row in data.iterrows():
    site_name = row['name']
    latitude, longitude, address = get_place_info(site_name)

    if latitude is not None and longitude is not None and address is not None:
        data.at[index, 'Latitud'] = latitude
        data.at[index, 'Longitud'] = longitude
        data.at[index, 'Dirección'] = address
    else:
        data.at[index, 'Latitud'] = "N/A"
        data.at[index, 'Longitud'] = "N/A"
        data.at[index, 'Dirección'] = "N/A"


data.to_csv("sitios_con_info.csv", index=False)
```

```python
#second try, more accurate + ratings
def get_place_info(place):

    try:
        geocode_result = gmaps.find_place(place, 'textquery', ['rating', 'formatted_address', 'geometry', 'price_level'])
    except Exception as e:
        print(f"Error al obtener la información para el sitio '{place}': {e}")
        return None

    if geocode_result:

        location = geocode_result[0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']
        address = geocode_result[0]['formatted_address']
        rating = geocode_result[0]['rating']
        try:
            price_level = geocode_result[0]['price_level']
        except:
            price_level = 'N/A'
        

        return latitude, longitude, address, rating, price_level
    else:
        print(f"No se encontró información para el sitio '{place}'.")
        return None, None, None, None, None


data = pd.read_excel("data_sites_english_2.xlsx")

data['Latitud'] = None
data['Longitud'] = None
data['Dirección'] = None
data['Rating'] = None
data['Precio'] = None


for index, row in data.iterrows():
    site_name = row['name'] + ' ' + row ['area']
    latitude, longitude, address, rating, price_level = get_place_info(site_name)

    if latitude is not None and longitude is not None and address is not None and rating is not None:
        data.at[index, 'Latitud'] = latitude
        data.at[index, 'Longitud'] = longitude
        data.at[index, 'Dirección'] = address
        data.at[index, 'Rating'] = rating
        data.at[index, 'Precio'] = price_level
    else:
        data.at[index, 'Latitud'] = "N/A"
        data.at[index, 'Longitud'] = "N/A"
        data.at[index, 'Dirección'] = "N/A"
        data.at[index, 'Rating'] = "N/A"
        data.at[index, 'Precio'] = "N/A"

data.to_csv("sitios_con_info2.csv", index=False)
```

```python
#third try, correcting the error
def get_place_info(place):
    try:
        geocode_result = gmaps.find_place(place, 'textquery', ['rating', 'formatted_address', 'geometry', 'price_level'])
    except Exception as e:
        print(f"Error al obtener la información para el sitio '{place}': {e}")
        return None, None, None, None, None

    if geocode_result.get('candidates', []):
        location = geocode_result['candidates'][0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']
        address = geocode_result['candidates'][0]['formatted_address']
        rating = geocode_result['candidates'][0].get('rating', 'N/A')
        price_level = geocode_result['candidates'][0].get('price_level', 'N/A')

        return latitude, longitude, address, rating, price_level
    else:
        print(f"No se encontró información para el sitio '{place}'.")
        return None, None, None, None, None


data = pd.read_excel("data_sites_english_2.xlsx")

data['Latitud'] = None
data['Longitud'] = None
data['Dirección'] = None
data['Rating'] = None
data['Precio'] = None

for index, row in data.iterrows():
    site_name = row['name'] + ' ' + row['area']
    latitude, longitude, address, rating, price_level = get_place_info(site_name)

    data.at[index, 'Latitud'] = latitude
    data.at[index, 'Longitud'] = longitude
    data.at[index, 'Dirección'] = address
    data.at[index, 'Rating'] = rating
    data.at[index, 'Precio'] = price_level

data.to_csv("sitios_con_info2.csv", index=False)
```
